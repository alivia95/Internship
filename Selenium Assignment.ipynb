```python
#scraping all the tags having experience
experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
#printing top 10 results
experience_tags[0:10]
```

```python
#displaying the tags into text
for i in experience_tags:
    experience=i.text
    experience_list.append(experience)
experience_list[0:10]
```

```python
#scraping all the tags having location
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
#printing top 10 results
location_tags[0:10]
```

```python
#displaying the tags into text
for i in location_tags:
    locations=i.text
    location.append(locations)
location[0:10]
```

```python
#printing the length of each list
print(len(job_titles),len(company_names),len(experience_list),len(location))
```

```python
#making a dataframe
jobs=pd.DataFrame({})
jobs['Title']=job_titles
jobs['company']=company_names
jobs['Experience Required']=experience_list
jobs['Location']=location

jobs[0:10]
```
