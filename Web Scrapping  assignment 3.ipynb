# 3. Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(2)

#fetching the url for access the webpage
driver.get("https://images.google.com/")
```

```python
#Searching for images with xpath
search_feild_product_name=driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
search_feild_product_name.send_keys('Fruits')

#to display the images for fruits in google
search_button1=driver.find_element_by_xpath('/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span')
search_button1.click()
```

```python
#getting images from website
for _ in range(20):
    driver.execute_script("window.scrollBy(0,100)")
images=driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')

img_urls=[]
img_data=[]
for image in images:
    source=image.get_attribute('src')
    if source is not None:
        if(source[0:4] == 'http'):
            img_urls.append(source)


#downloading the images
for i in range(len(img_urls)):
    if i >10:
        break
    print("Downloading {0} for {1} images" .format(i, 10))
    response= requests.get(img_urls[i])
    file = open(r"C:\Users\alivi\Downloads\images"+str(i)+"jpg","wb")
    file.write(response.content)

```
